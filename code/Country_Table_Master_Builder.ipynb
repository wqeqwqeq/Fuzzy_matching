```python
# Method 2
from googlesearch import search
def getWebPair(table):
    name_list = table['company'].unique().tolist()
    urls = []
    for i, company in enumerate(name_list):
        print('testing: ' + company)
        for j in search(company, tld="com", num=10, stop=1, pause=3): 
            print( str(i) +'. '+company+': '+j)
            urls.append([company,j]);
    df_pair = pd.DataFrame(urls)
    return df_pair
```

```python
# inner join 
def merge_all_url_inner(table_list):
    df = pd.DataFrame(columns = ['company_url'])
    for i, table in enumerate(table_list) :
        if(i == 0):
            df = df.merge(table, on = 'company_url', how = 'outer')
        else:
            df = df.merge(table, on = 'company_url', how = 'inner')
        print('rows: ' + str(len(df)))
    return df
```

```python
# merging function on company website(url)
def merge_all_url(table_list):
    df = pd.DataFrame(columns = ['company_url'])
    for table in table_list :
        df = df.merge(table, on = 'company_url', how = 'outer')
        print('rows: ' + str(len(df)))
    return df
```

```python
# merging function
def merge_all(table_list):
    df = pd.DataFrame(columns = ['company'])
    for table in table_list :
        df = df.merge(table, on = 'company', how = 'outer')
        print('rows: ' + str(len(df)))
    return df
```
